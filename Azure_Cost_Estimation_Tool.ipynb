```python
vm_hourly_cost = 0.05
storage_cost_per_gb = 0.02
bandwidth_cost_per_gb = 0.01
print("Azure Cost Estimation Tool\n")
vm_hours = float(input("Enter VM usage hours: "))
storage_gb = float(input("Enter Storage usage (GB): "))
bandwidth_gb = float(input("Enter Bandwidth usage (GB): "))
vm_cost = vm_hours * vm_hourly_cost
storage_cost = storage_gb * storage_cost_per_gb
bandwidth_cost = bandwidth_gb * bandwidth_cost_per_gb
total_cost = vm_cost + storage_cost + bandwidth_cost
print("\nEstimated Azure Cost Details")
print("VM Cost: $", round(vm_cost, 2))
print("Storage Cost: $", round(storage_cost, 2))
print("Bandwidth Cost: $", round(bandwidth_cost, 2))
print("Total Estimated Cost: $", round(total_cost, 2))
```
