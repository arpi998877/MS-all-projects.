```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Student Name": ["Amit", "Riya", "Karan", "Sneha", "Rahul"],
    "Maths": [78, 85, 62, 90, 45],
    "Science": [72, 88, 55, 92, 40],
    "English": [80, 79, 60, 85, 50],
    "Computer": [90, 95, 70, 98, 65]
}

df = pd.DataFrame(data)

df["Total Marks"] = df.iloc[:, 1:5].sum(axis=1)

df["Percentage"] = df["Total Marks"] / 4

def assign_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    else:
        return "D"

df["Grade"] = df["Percentage"].apply(assign_grade)

df["Result"] = df["Percentage"].apply(lambda x: "Pass" if x >= 40 else "Fail")

print(df)

plt.figure()
subject_avg = df.iloc[:, 1:5].mean()
subject_avg.plot(kind="bar")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Subject-wise Average Marks")
plt.show()

plt.figure()
result_count = df["Result"].value_counts()
result_count.plot(kind="pie", autopct="%1.1f%%")
plt.title("Pass vs Fail Percentage")
plt.ylabel("")
plt.show()
```
